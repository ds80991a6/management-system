```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import ipywidgets as widgets
from IPython.display import display

# Load movies data
movies = pd.read_csv("movies.csv")

# Function to clean movie titles
def clean_title(title):
    return re.sub("[^a-zA-Z0-9 ]", "", title)

# Apply cleaning function to create a new column
movies["clean_title"] = movies["title"].apply(clean_title)

# Vectorize movie titles
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
tfidf = vectorizer.fit_transform(movies["clean_title"])

# Function to search for similar movies
def search(title):
    title = clean_title(title)
    query_vec = vectorizer.transform([title])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity, -5)[-5:]
    results = movies.iloc[indices][::-1]
    return results

# Define movie input widget
movie_input = widgets.Text(
    value="toy story",
    description="Movie Title:",
    disabled=False
)

# Define movie list output widget
movie_list = widgets.Output()

# Define function to handle input changes
def on_type(data):
    with movie_list:
        movie_list.clear_output()
        title = data["new"]
        if len(title) > 5:
            display(search(title))

# Attach event handler to input widget
movie_input.observe(on_type, names='value')

# Display widgets
display(movie_input, movie_list)

# Load ratings data
ratings = pd.read_csv("ratings.csv")

# Function to find similar movies
def find_similar_movies(movie_id):
    similar_users = ratings[(ratings["movieId"] == movie_id) & (ratings["rating"] > 5)]["userId"].unique()
    similar_user_recs = ratings[(ratings["userId"].isin(similar_users)) & (ratings["rating"] > 4)]["movieId"]

    similar_user_recs = similar_user_recs.value_counts()
    similar_user_recs = similar_user_recs[similar_user_recs > float(0.1)]

    all_users = ratings[(ratings["movieId"].isin(similar_user_recs.index)) & (ratings["rating"] > 4)]
    all_users_recs = len(all_users["userId"].unique())

    rec_percentages = pd.DataFrame({
        "similar": similar_user_recs,
        "all": [all_users_recs]
    })

    rec_percentages["score"] = rec_percentages["similar"] / rec_percentages["all"]

    rec_percentages = rec_percentages.sort_values("score", ascending=False)
    
    # Merge with movies data
    recommendations = rec_percentages.head(10).merge(movies, left_index=True, right_on="movieId")[["score", "title", "genres"]]
    return recommendations

# Define movie name input widget
movie_name_input = widgets.Text(
    value="Toy Story",
    description="Movie Title:",
    disabled=False
)

# Define recommendation list output widget
recommendation_list = widgets.Output()

# Define function to handle input changes for movie name input
def on_type(data):
    with recommendation_list:
        recommendation_list.clear_output()
        title = data["new"]
        if len(title) > 5:
            results = search(title)
            if not results.empty:
                movie_id = results.iloc[0]["movieId"]
                display(find_similar_movies(movie_id))
            else:
                print("No similar movies found.")

# Attach event handler to movie name input widget
movie_name_input.observe(on_type, names="value")

# Display movie name input and recommendation list widgets
display(movie_name_input, recommendation_list)
```
